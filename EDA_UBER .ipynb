#### Chart - 5

````python
#Prepare data
line_data = df.groupby(['Request Day', 'Request Hour'])['Request id'].count().reset_index()

# Ensure correct day order
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
line_data['Request Day'] = pd.Categorical(line_data['Request Day'], categories=day_order, ordered=True)

# Plot Facet Grid
g = sns.FacetGrid(line_data, col='Request Day', col_wrap=3, height=4, sharey=False)
g.map(sns.lineplot, 'Request Hour', 'Request id', marker='o', color='#4CAF50')
g.set_axis_labels("Hour of the Day", "Number of Requests")
g.set_titles("{col_name}")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Requests per Hour by Day of the Week')
plt.show()
````

I picked a Facet Grid line chart to avoid clutter and provide clear, individual trends for each day, making the hourly demand pattern easy to interpret.

This chart shows daily variations in peak request hours, highlighting which days have morning peaks, evening peaks, or uniform demand.

##### 3. Will the gained insights help creating a positive business impact?

#### Chart - 8

````python
# Chart - 8 visualization code
# Converting timestamps to datetime
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'], dayfirst=True, errors='coerce')
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'], dayfirst=True, errors='coerce')

# Calculating trip duration in minutes for completed trips only
df['Trip Duration (min)'] = (df['Drop timestamp'] - df['Request timestamp']).dt.total_seconds() / 60

# Filtering completed trips
completed_trips = df[df['Status'] == 'Trip Completed']

# Plotting histogram
plt.figure(figsize=(8,5))
completed_trips['Trip Duration (min)'].plot(kind='hist', bins=30, color='#9C27B0')
plt.title('Distribution of Trip Durations (in minutes)')
plt.xlabel('Trip Duration (min)')
plt.ylabel('Number of Trips')
plt.show()
````

##### 1. Why did you pick the specific chart?

I picked a histogram to visualise the distribution of trip durations, identifying common trip lengths.

##### 2. What is/are the insight(s) found from the chart?

This chart shows which trip durations are most frequent, indicating whether trips are short commutes or long-distance airport transfers.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, insights help optimise pricing strategy, driver shift planning, and fuel cost estimation (positive impact). Ignoring trip duration patterns can lead to inefficient resource allocation and poor profitability management (negative impact).
